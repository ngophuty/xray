```python
import numpy as np, pandas as pd
from glob import glob
import shutil, os
import matplotlib.pyplot as plt
from sklearn.model_selection import GroupKFold
from tqdm.notebook import tqdm
import seaborn as sns
fold = 0
train_df = pd.read_csv('/content/drive/Shareddrives/yolov5_5fold/train.csv')
train_df.head()
```

```python
train_df = train_df[train_df.class_id!=14].reset_index(drop = True)
```

```python
train_df['image_path'] = f'/content/drive/Shareddrives/yolov5_5fold/data/images/'+train_df.image_id+('.jpg')
train_df.head()
```

```python
gkf  = GroupKFold(n_splits = 5)
train_df['fold'] = -1
for fold, (train_idx, val_idx) in enumerate(gkf.split(train_df, groups = train_df.image_id.tolist())):
    train_df.loc[val_idx, 'fold'] = fold
train_df.head()
#train_df.to_csv('/content/drive/Shareddrives/yolov5_5fold/data/train_fold.csv')
```

```python
%cd /content/drive/Shareddrives/yolov5_5fold
import pandas as pd
import numpy as np, pandas as pd
from glob import glob
import shutil, os
import matplotlib.pyplot as plt
from sklearn.model_selection import GroupKFold
from tqdm.notebook import tqdm
import seaborn as sns
fold =3
train_df=pd.read_csv('/content/drive/Shareddrives/yolov5_5fold/data/train_fold.csv')
train_df.head()
```

```python
train_files = []
val_files   = []
val_files += list(train_df[train_df.fold==fold].image_path.unique())
train_files += list(train_df[train_df.fold!=fold].image_path.unique())
len(train_files), len(val_files)
#print('train data fold',fold)
```

```python
label_dir = '/content/drive/Shareddrives/yolov5_5fold/data/labels'
for file in tqdm(train_files):
    shutil.copy(file, '/content/drive/Shareddrives/yolov5_5fold/data/fold_3/images/train')
    filename = file.split('/')[-1].split('.')[0]
    shutil.copy(os.path.join(label_dir, filename+'.txt'), '/content/drive/Shareddrives/yolov5_5fold/data/fold_3/labels/train')
```

```python
label_dir = '/content/drive/Shareddrives/yolov5_5fold/data/labels'
for file in tqdm(val_files):
    shutil.copy(file, '/content/drive/Shareddrives/yolov5_5fold/data/fold_3/images/val')
    filename = file.split('/')[-1].split('.')[0]
    shutil.copy(os.path.join(label_dir, filename+'.txt'), '/content/drive/Shareddrives/yolov5_5fold/data/fold_3/labels/val')
```
