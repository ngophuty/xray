```python
import os 
import numpy as np
import pandas as pd
def nomaz(w,h,box):
  x=box[:,[2,4]]*w
  y=box[:,[3,5]]*h
  xmin = x[:,[0]]-x[:,[1]]/2
  xmax = x[:,[0]]+x[:,[1]]/2
  ymin = y[:,[0]]-y[:,[1]]/2
  ymax = y[:,[0]]+y[:,[1]]/2
  box[:,[2]]=xmin
  box[:,[3]]=ymin
  box[:,[4]]=xmax
  box[:,[5]]=ymax
  return box
df = pd.read_csv('/content/drive/Shareddrives/trainmodel/yolov5_WBF/yolov5/runs/detect/test.csv')
path ='/content/drive/Shareddrives/yolov5_5fold/yolov5/runs/detect/detect_fold_4_TTA/labels/'
image_ids = []
PredictionStrings = []
for file1 in os.listdir(path):
   file_name = os.path.splitext(file1)[0]
   w,h =df.loc[df.image_id==file_name,['width','height']].values[0]
   f = open(path+file1,'r')
   a= f.read().replace('\n',' ').strip().split(' ')
   new=[]
   for item in a:
     new.append(float(item))
   new =np.array(new).astype(np.float32).reshape(-1,6)
   box=new[:,[0,5,1,2,3,4]]
   gh=nomaz(w,h,box)
   data=np.round(gh[:],3)
   data =list(data.reshape(-1).astype(str))
   for idx in range(len(data)):
     data[idx]=str(int(float(data[idx]))) if idx%6!=1 else data[idx]
   image_ids.append(file_name)
   PredictionStrings.append(' '.join(data))
pred_df = pd.DataFrame({'image_id':image_ids,
                        'PredictionString':PredictionStrings})
sub_df = pd.merge(df, pred_df, on = 'image_id', how = 'left').fillna("14 1.0 0 0 1 1")
sub_df = sub_df[['image_id', 'PredictionString']]
sub_df.to_csv('/content/drive/Shareddrives/yolov5_5fold/yolov5/runs/yolov5_fold_4_TTA.csv',index = False)
sub_df.tail()
```
